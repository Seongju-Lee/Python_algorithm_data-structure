```python
# 초간단 해쉬 함수 만들어보기.
# 가장 간단한 방법: Division방법(나누기를 통한 나머지 값을 사용하는 기법)

# 어떠한 키든 간에 5로 나누어 나머지를 출력하기 때문에 고정된길이로 나온다.

def hash_func(key):
    return key % 5

# 해쉬 테이블에 데이터 저장

# 데이터에 따라 필요시 key생성 방법 정의가 필요함
data1 = 'Andy'
data2 = 'Dave'
data3 = 'Trump'

# 문자의 아스키 코드값 추출

# 데이터의 키값 추출: ord(data[0])
# 추출된 키값을 해쉬함수에 넣어줌: hash_func
# 리턴값은 해쉬 주소, 해쉬값이 됨
print(ord(data1[0]),ord(data2[0]),ord(data3[0]))
print(ord(data1[0]),hash_func(ord(data2[0])),ord(data3[0]))
```

```python
# 해쉬 테이블에 값 저장

def storage_data(data, value):
    key = ord(data[0]) # 데이터의 키 값을 만듬
    hash_addr = hash_func(key) # 키값을 이용하여 해쉬함수에 적용 -> 해쉬 주소 받아옴

    hash_table[hash_addr] = value # 해쉬 테이블의 해당 주소에 데이터 넣어줌
```

```python
storage_data('Andy', '01012345678')
storage_data('Dave', '01022225678')
storage_data('Trump', '01044445678')

# 실제 데이터 저장하고, 읽어오기

def get_data(data):
    key = ord(data[0])
    hash_addr = hash_func(key)

    return hash_table[hash_addr]

get_data('Andy')
```

```python
# 연습1: 리스트 변수를 활용해서 해쉬 테이블 구현해보기
# 1. 해쉬 함수: key % 8
# 2. 해쉬 키 생성: hash(data)

hash_table = list([0 for i in range(8)])

def get_key(data):
    return hash(data)

def hash_func(key):
    return key % 8


def set_data(data, value):
    hash_addr = hash_func(get_key(data))
    hash_table[hash_addr] = value

def get_value(data):
    hash_addr = hash_func(get_key(data))
    return hash_table[hash_addr]
```

```python
set_data('Dave', '01011112222')
set_data('Andy', '01033334444')
print(get_value('Dave'))
print(get_value('Andy'))
hash_table
```

```python
# 충돌 문제 해결

# 1. Open Hasing중 하나인 chaining 기법
#     -> 서로 다른 데이터가 해쉬 어드레스가 충돌이 일어나면, 해당 어드레스의 값을 링크드 리스트로 연결시키는 기법

hash_table = list([0 for _ in range(8)])

def get_key(data):
    return hash(data)

def hash_func(key):
    return key % 8

def save_data(data, value):
    index_key = get_key(data) # data로부터 얻어진(hash(data)) 키 값을 저장하는 이유는
                              # 해쉬 값 안에 데이터가 두 개 이상일 경우 어떠한 데이터가 원하는 값인지 구별하기 위해
                              # slot에 같이 저장을 해놓기 위해서이다.

    hash_addr = hash_func(get_key(data))
    if hash_table[hash_addr] != 0:
        for i in range(len(hash_table[hash_addr])):
            if hash_table[hash_addr][i][0] == index_key:
                hash_table[hash_addr][i][1] = value
                return
        hash_table[hash_addr].append([index_key, value]) 
    
    else:
        hash_table[hash_addr] = [[index_key, value]]

def read_data(data):
    index_key = get_key(data)
    hash_addr = hash_func(get_key(data))

    if hash_table[hash_addr] != 0:
        for i in range(len(hash_table[hash_addr])):
            if hash_table[hash_addr][i][0] == index_key:
                return hash_table[hash_addr][i][1]
        return None
    else:
        return None
```

```python
save_data('Data', '123456789')
save_data('Ddqwe', '432432432')

read_data('Ddd')
```

```python
# 1. Close Hasing중 하나인 Linear Probing 기법
#     -> 서로 다른 데이터가 해쉬 어드레스가 충돌이 일어나면, 하나씩 비교해 나아가며, 빈 슬롯에 데이터를 삽입하는 방식
#        저장공간 활용도를 높일 수 있다.

hash_table = list([0 for _ in range(8)])

def get_key(data):
    return hash(data)

def hash_func(key):
    return key % 8

def save_data(data, value):
    index_key = get_key(data)

    hash_addr = hash_func(get_key(data))
    if hash_table[hash_addr] != 0: # 해당 hash address에 데이터가 저장되어 있음
        for i in range(hash_addr, len(hash_table)): # 데이터로부터 나온 hash address부터 끝까지 순회
            if hash_table[i] == 0:
                hash_table[i] = [index_key, value]
                return
            elif hash_table[i][0] == index_key: # 동일한 키를 가진 데이터가 존재한 경우, 데이터를 수정하는 부분
                hash_table[i][1] = value
                return
    else:
        hash_table[hash_addr] = [index_key, value]


def read_data(data):
    index_key = get_key(data)
    hash_addr = hash_func(index_key)

    if hash_table[hash_addr] != 0:
        for i in range(hash_addr, len(hash_table)):
            if hash_table[i] == 0:
                return None
            elif hash_table[i][0] == index_key:
                return hash_table[i][1]
    else:
        return None
```

```python
save_data('de', '01200123123')
save_data('dw', '3333333333')
read_data('dw')

```
