```python
# Quadrangle 클래스를 만들고, 직사각형 1개 객체와 정사각형 1개 객체를 생성하고, 각 사각형의 넓이를 출력하기.abs
# 직사각형 2개 속성: width = 10, height = 5, color = 'red'
# 직사각형 2개 속성: width = 7, height = 7, color = 'blue'
class Quadrangle:
    width = 0
    height = 0
    color = 'black'

    # 파이썬의 메소드의 첫 인자는 항상 self가 들어가므로 항상 명시해줘야함.
    def get_area(self):
        return self.width * self.height

    def set_area(self, height, width):
        self.height = height
        self.width = width


square1 = Quadrangle() # 직사각형
square2 = Quadrangle() # 정사각형

square1.set_area(10,5)
square2.set_area(7,7)

square1.color = 'red'
square2.color = 'blue'


print(square1.get_area())
print(square2.get_area())
print('직사각형의 색: ', square1.color)
print('정사각형의 색: ', square2.color)
```

```python
# Linked_List 구현

class Node:

    def __init__(self, data, next=None): # 파이썬 클래스의 생성자: __init__, 소멸자: __del__
        self.data = data
        self.next = next


```

```python
node1 = Node(1)
node2 = Node(2)

node1.next = node2
head = node1 # 가장 앞의 노드를 알기 위함.
```

```python
# 링크드 리스트로 데이터 추가하기

class Node:

    def __init__(self,data,next = None):
        self.data = data
        self.next = next

    
def add(data):
    node = head
    while node.next:
        node = node.next
    node.next = Node(data)

```

```python
node1 = Node(1)
node2 = Node(2)

node1.next = node2
head = node1

for i in range(3,11):
    add(i)




```

```python
# 파이썬 객체지향 프로그래밍으로 링크드 리스트 구현하기(체계적)

class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

class NodeMgmt:
    
    def __init__(self, data):
        self.head = Node(data)

    def add(self, data):
        if self.head == None:
            self.head = Node(data)
        else:
            node = self.head
            while node.next:
                node = node.next
            
            node.next = Node(data)

    def desc(self): # 링크드리스트를 순회 하면서 존재하는 데이터를 가져오는 것
        node = self.head
        while node:
            print(node.data)
            node = node.next


```

```python
linked_list1 = NodeMgmt(0)
linked_list1.desc()
```
