```python
def sum_all(n):
    total = 0

    for num in range(1,n+1):
        total += num
    return total        # 결과적으로 for문이 n번 동작하므로 O(n)이라고 할 수 있음
```

```python
sum_all(100) 
```

```python
def sum_all(n):
    return n*(n+1) // 2 # n이 몇번을 들어오든 한번 수행하게 된다. 결과적으로 O(1)이라고 할 수 있음.
```

```python
sum_all(100)
```
