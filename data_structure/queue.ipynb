```python
# 어디에 큐가 많이 쓰일까?
# 멀티 태스킹을 위한 프로세스 스케쥴링 방식을 구현하기 위해 많이 사용됨.(운영체제 참조)

# ex1 리스트 변수로 큐를 다루는 인큐, 디큐 기능 구현해보기
```

```python
queue_list = list()

def enqueue(data):
    queue_list.append(data)

def dequeue():
    data = queue_list[0]
    del(queue_list[0])
    return data
```

```python
for i in range(10):
    enqueue(i)
```

```python
len(queue_list)
```

```python
dequeue() # 하나씩 해보면 앞에서부터 빠질 것. FIFO 구조
```
