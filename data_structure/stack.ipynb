```python
# 리스트 변수로 스택을 다루는 pop, push 기능 구현해보기. 
stack_list = list()
def push(data):
    stack_list.append(data)
def pop():
    data = stack_list[-1] # 리스트 인덱스에 -1은 크기에 상관없이 마지막 인덱스의 값을 가져오는 것을 의미한다.
    del stack_list[-1]
    return data
```

```python
for i in range(10):
    push(i)
```
